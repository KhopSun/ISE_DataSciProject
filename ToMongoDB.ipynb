### From local files to MongoDB

```python
db = client['DsdeDatabase']  # Replace with your database name
```

```python
# Define the base path of the folders
# Replace with the actual path to your folders (2018-2023)
def add_to_mongo(folder_name, collection_name):
    collection = db[collection_name]
    for file in os.listdir(folder_name): # name of folder in the same directory
        with open(os.path.join(folder_name, file), 'r') as f:
            data = json.load(f)
        f.close()
        collection.insert_one(data)
        print(f'finish inserting {file}')
    print("Data insertion complete!")
```

```python
add_to_mongo('SelectedData', 'data_given')
```
